```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Function to generate a synthetic dataset
def generate_dataset(size=100, noise_level=20):
    np.random.seed(42)
    black_data = np.random.normal(50, noise_level, 20)  # Black readings
    white_data = np.random.normal(200, noise_level, 80)  # White readings
    dataset = np.concatenate((black_data, white_data))
    np.random.shuffle(dataset)
    return dataset
```

```python
# Function to perform k-means clustering
def classify_data_with_kmeans(data, n_clusters=2):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data.reshape(-1, 1))
    centroids = kmeans.cluster_centers_
    return labels, centroids
```

```python
# Function to plot data and centroids
def plot_data_with_centroids_and_save(data, labels, centroids, fig_size=(10, 6)):
    plt.figure(figsize=fig_size)
    unique_labels = np.unique(labels)
    for label in unique_labels:
        plt.scatter(data[labels == label], [label] * len(data[labels == label]), label=f'Cluster {label}')
    plt.scatter(centroids, [0.2] * len(centroids), color='red', marker='x', s=100, label='Centroids')
    plt.title("IR Sensor Data Classification")
    plt.xlabel("Sensor Readings")
    plt.ylabel("Clusters")
    plt.legend()
    plt.show()
```

```python
# Generate dataset
dataset = generate_dataset(size=100, noise_level=10)

# Classify data using k-means
labels, centroids = classify_data_with_kmeans(dataset)
print("Number of clusters: ", labels.shape)
print("Number of points in cluster 0: ", np.size(np.where(labels == 0)))
print("Number of points in cluster 1: ", np.size(np.where(labels == 1)))
print("Centroid of cluster 0: ", centroids[0])
print("Centroid of cluster 1: ", centroids[1])
# Plot data and centroids
plot_data_with_centroids_and_save(dataset, labels, centroids, fig_size=(8, 5))

# Display dataset and centroids for verification
# dataset, centroids.ravel()
```
